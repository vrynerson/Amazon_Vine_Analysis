```python
# import dependencies
import pandas as pd
```

```python
# read in csv and make df

df = pd.read_csv('vine_table.csv')
df.head(10)
```

```python
# drop null information rows
df = df.dropna()
df.head(10)
```

```python
# total vine reviews
total_vine = df[df['vine'] == 'Y']
total_vine['vine'].count()
```

```python
# not vine
total_not_vine = df[df['vine'] == 'N']
total_not_vine['vine'].count()
```

```python
#filter data and create new df to retrieve rows where total_votes is >=20

filter_tv_twenty_df = df[df['total_votes'] >=20.0]
filter_tv_twenty_df.head(10)
```

```python
# filter filter_tv_twenty_df to retrieve all rows where number of "helpful_votes"/"total_votes">=50%
helpful_reviews_df = filter_tv_twenty_df[filter_tv_twenty_df["helpful_votes"]/filter_tv_twenty_df["total_votes"]>=0.5]
helpful_reviews_df
```

```python
# filter helpful_reviews_df to retrieve all rows where a review was written as part of vine program 
# vine == 'Y'

yes_vine_df = helpful_reviews_df[helpful_reviews_df['vine'] == 'Y']
yes_vine_df
```

```python
# 5 star yes_vine_df
five_star_yes_vine_df = yes_vine_df[yes_vine_df['star_rating'] == 5.0]
five_star_yes_vine_df
```

```python
# filter helpful_reviews_df to retrieve all rows where a review was NOT written as part of vine program 
# vine == 'N'

no_vine_df = helpful_reviews_df[helpful_reviews_df['vine'] == 'N']
no_vine_df
```

```python
# 5 star no_vine_df
five_star_no_vine_df = no_vine_df[no_vine_df['star_rating'] == 5.0]
five_star_no_vine_df
```

```python
# total number of reviews

total_reviews_count = df['review_id'].count()
total_reviews_count
```

```python
# total number of helpful reviews

total_helpful_reviews_count = helpful_reviews_df['review_id'].count()
total_helpful_reviews_count
```

```python
#five_star_review df
five_star_review = helpful_reviews_df.loc[helpful_reviews_df['star_rating'] == 5.0]
five_star_review
```

```python
#total number of 5 star reviews
five_star_count = five_star_review['star_rating'].count()
f"Number of five star reviews for pet products: {five_star_count}"
```

```python
# 5 star paid reviews 
five_star_paid = five_star_review[five_star_review['vine'] == 'Y']
five_star_paid_count = five_star_paid['star_rating'].count()
f"Number of helpful Vine program 5 star reviews: {five_star_paid_count}" 
```

```python
# percentage of 5 star paid reviews (out of all reviews)
percentage_paid_five_star = five_star_paid_count / total_reviews_count * 100 
f"{percentage_paid_five_star:.2f}% of the reviews on pet products are part of the paid Vine program."
```

```python
# percentage of 5 star paid reviews (out of 5 star reviews)
percentage_paid_five_star_out_of_five_star = five_star_paid_count / five_star_count * 100 
f"{percentage_paid_five_star_out_of_five_star:.2f}% of the helpful five star reviews on pet products are part of the paid Vine program."
```

```python
# 5 star unpaid reviews 
five_star_unpaid = five_star_review[five_star_review['vine'] == 'N']
five_star_unpaid_count = five_star_unpaid['star_rating'].count()
f"Number of helpful non-Vine program 5 star reviews: {five_star_unpaid_count}" 
```

```python
# percentage of 5 star unpaid reviews (out of all reviews)
percentage_unpaid_five_star = five_star_unpaid_count / total_reviews_count * 100 
f"{percentage_unpaid_five_star:.2f}% of the helpful reviews on pet products are honest reviews."
```

```python
# percentage of 5 star paid reviews (out of 5 star reviews)
percentage_paid_five_star_out_of_five_star = five_star_paid_count / five_star_count * 100 
print(f"{percentage_paid_five_star_out_of_five_star:.2f}% of the helpful five star reviews on pet products are part of the paid Vine program.")
```

```python
percentage_paid_five_star_out_of_five_star + percentage_unpaid_five_star_out_of_five_star
```

```python
print(f"Number of five star reviews for pet products: {five_star_count}")
print(f"Number of helpful non-Vine program 5 star reviews: {five_star_unpaid_count}")
print(f"Number of helpful Vine program 5 star reviews: {five_star_paid_count}")
print(f"{percentage_unpaid_five_star_out_of_five_star:.2f}% of the helpful five star reviews on pet products are not part of the Vine program.")
print(f"{percentage_paid_five_star_out_of_five_star:.2f}% of the helpful five star reviews on pet products are part of the paid Vine program.")
```
