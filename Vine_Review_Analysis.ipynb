```python
# import dependencies
import pandas as pd
```

```python
# read in csv and make df

df = pd.read_csv('vine_table.csv')
df.head(10)
```

```python
# drop null information rows
df = df.dropna()
df.head(10)
```

```python
#filter data and create new df to retrieve rows where total_votes is >=20

filter_tv_twenty_df = df[df['total_votes'] >=20.0]
filter_tv_twenty_df.head(10)
```

```python
# filter filtered_df to retrieve all rows where number of "helpful_votes"/"total_votes">=50%

filter_fifty_percent_votes_df = filter_tv_twenty_df[filter_tv_twenty_df["helpful_votes"]/filter_tv_twenty_df["total_votes"]>=0.5]
filter_fifty_percent_votes_df.head(10)
```

```python
# filter filtered1_df to retrieve all rows where a review was written as part of vine program 
# vine == 'Y'

yes_vine_df = filter_fifty_percent_votes_df[filter_fifty_percent_votes_df['vine'] == 'Y']
yes_vine_df.head(10)
```

```python
# filter filtered1_df to retrieve all rows where a review was NOT written as part of vine program 
# vine == 'N'

no_vine_df = filter_fifty_percent_votes_df[filter_fifty_percent_votes_df['vine'] == 'N']
no_vine_df.head(10)
```

```python
# total number of reviews

total_reviews_count = df['review_id'].count()
total_reviews_count
```

```python
#five_star_review df
five_star_review = df.loc[df['star_rating'] == 5.0]
five_star_review.head(10)
```

```python
#total number of 5 star reviews
five_star_count = five_star_review['star_rating'].count()
five_star_count
```

```python
# 5 star paid reviews 
five_star_paid = five_star_review[five_star_review['vine'] == 'Y']
five_star_paid_count = five_star_paid['star_rating'].count()
five_star_paid_count
```

```python
# percentage of 5 star paid reviews (out of all reviews)
percentage_paid_five_star = five_star_paid_count / total_reviews_count * 100 
f"{percentage_paid_five_star:.2f}% of the reviews on pet products are part of the paid Vine program."
```

```python
# percentage of 5 star paid reviews (out of 5 star reviews)
percentage_paid_five_star_out_of_five_star = five_star_paid_count / five_star_count * 100 
f"{percentage_paid_five_star_out_of_five_star:.2f}% of the five star reviews on pet products are part of the paid Vine program."
```

```python
# 5 star unpaid reviews 
five_star_unpaid = five_star_review[five_star_review['vine'] == 'N']
five_star_unpaid_count = five_star_unpaid['star_rating'].count()
five_star_unpaid_count 
```

```python
# percentage of 5 star unpaid reviews (out of all reviews)
percentage_unpaid_five_star = five_star_unpaid_count / total_reviews_count * 100 
f"{percentage_unpaid_five_star:.2f}% of the reviews on pet products are honest reviews."
```

```python
# percentage of 5 star paid reviews (out of 5 star reviews)
percentage_unpaid_five_star_out_of_five_star = five_star_unpaid_count / five_star_count * 100 
f"{percentage_unpaid_five_star_out_of_five_star:.2f}% of the five star reviews on pet products are honest reviews."
```
